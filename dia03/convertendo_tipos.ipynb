Importar CSV

```python
import pandas as pd

df = pd.read_csv("../data/customers.csv", sep=";")
df
```

Criar nova Serie(Coluna) de Points com valores aumentados em 2x

```python

df["Points_dobble"] = df["Points"] * 2
df
```

Mostrar o Dataframe com a coluna Points_dobble convertida para float64

```python

from numpy import float64


df[["Points", "Points_dobble"]].astype(float64)
```

Criar uma nova coluna "Lista" usando for para injetar em i rows
