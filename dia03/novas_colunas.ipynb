Pega todos os i rows da serie de Name e cria um novo campo Nome_Maiusculo com os valores copiados de Name em maiusculo

```python
nomes_maiusculo = []
for i in df['Name']:
    nomes_maiusculo.append(i.upper())

df["Nome_Maiusculo"] = nomes_maiusculo
df
```

Cria a coluna name_first, usando a função get_first() que aplica o algoritmo do método criado para os valores da serie de Name

```python
def get_first(nome:str):
    nome = nome.upper()
    return nome.split("_")[0]

df["Name_First"] = df["Name"].apply( get_first )
df
```

Aplica um Upper Case nas rows da serie de Name e printa (As alterações não são persistidas).

Cria coluna Faixa_Pontos aplicando o metodo intervalo_pontos na coluna de Points

```python

def intervalo_pontos(pontos):
    if pontos < 2500:
        return "baixo"
    elif pontos < 3500:
        return "medio"
    else:
        return "alto"
    
df["Faixa_Pontos"] = df["Points"].apply(intervalo_pontos)
df
```

Criar o DataFrame df_crm usando o dicionário data

```python

data = {
    "nome": ["Teo", "Nah", "Maria", "Lara"],
    "recencia": [1,30,10,45],
    "valor":[100,2000, 15, 500],
    "frequencia":[2, 5, 1, 15]
}

df_crm = pd.DataFrame(data)
df_crm
```

Criar coluna RFV usando condições em cima das colunas especificadas dentro do metodo rfv

```python

def rfv(row):
    
    nota = 0
    
    if row['recencia'] <= 10:
        nota += 10
    elif 10 < row['recencia'] <= 30:
        nota += 5
    elif row['recencia'] > 30:
        nota += 0

    if row['valor'] > 1000:
        nota += 10
    elif 100 <= row['valor'] < 1000:
        nota += 5
    elif row['valor'] < 100:
        nota += 0

    if row['frequencia'] > 10:
        nota += 10
    elif 5 <= row['frequencia'] < 10:
        nota += 5
    elif row['frequencia'] < 5:
        nota += 0
    
    return nota

df_crm["RFV"] = df_crm.apply(rfv, axis=1)
df_crm
```
