Importação de CSV

```python
import pandas as pd

csv = pd.read_csv("../data/customers.csv", sep=";")

#Número de colunas
len(csv.columns)

#Número de linhas
len(csv.index)

#Todos os UUIDs como Serie
csv.UUID

#Shape com valores RowsxColumns numa tupla
csv.shape


```

Condicionais básicas no Pandas


```python

maxPoints = csv["Points"].max()
conditionMax = csv["Points"] == maxPoints

condition = csv["Points"] > 2000

conditionMore1000Less2000 =  (csv["Points"] >= 1000) & (csv["Points"] < 2000)
df_1000_2000 = csv[conditionMore1000Less2000].copy()

#Manipulando uma cópia dos dados para não mexer direto na referencia
df_1000_2000["Points"] = df_1000_2000["Points"] + 1000

csv[condition]
csv[conditionMax]
df_1000_2000

```

Ordenação

```python
csv = (csv.sort_values( by=["Points", "Name"],
                      ascending=[False, True] )
        .rename(columns={"Name":"Nome", "Points":"Pontos"}))

csv
```
