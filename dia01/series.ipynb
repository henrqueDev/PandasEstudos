```python
import math
import pandas as pd # type: ignore

idades = [1,2,3,4,5,10]

series_idades = pd.Series(idades, name="Idades")
```

```python
#Contagem de elementos
series_idades.count()

#  Média
series_idades.mean()

# Variancia
series_idades.var()

# Desvio Padrão
series_idades.std()

#Mediana
series_idades.median()

# Descrever todas as informações do series
series_idades.describe()

```

```python
# Propriedade Shape (Tupla)
series_idades.shape

series_idades[0]

# Alterando index da serie
series_idades.index = [20, 30, 50, 60, 30, 40]

# Navegando nos indices novos
#series_idades['b']

# Pegar dados das posições como lista (Independe dos indices)
# Usar quando a ordem for relevante
series_idades.iloc[0:]

# Pegar dados das posições como Dicionario
series_idades.loc[60]
```
